```python
import cv2
import matplotlib.pyplot as plt
Gray_image = cv2.imread('kookieg.jpeg')
plt.imshow(Gray_image)
plt.show()
hist = cv2.calcHist([Gray_image],[0],None,[256],[0,256])
plt.figure()
plt.title("Histogram")
plt.xlabel('Grayscale_value')
plt.ylabel('pixel count')
plt.stem(hist)
plt.show()


```

```python
Color_image = cv2.imread('kookie.jpeg')
plt.imshow(Color_image)
plt.show()
hist1 = cv2.calcHist([Color_image],[0],None,[256],[0,256])
plt.figure()
plt.title("Histogram")
plt.xlabel('Intensity_value')
plt.ylabel('pixel count')
plt.stem(hist1)
plt.show()
```

```python
import cv2
Gray_image=cv2.imread('kookie.jpeg',0)
equ=cv2.equalizeHist(Gray_image)
cv2.imshow('Gray Image',Gray_image)
cv2.imshow('Equalized Image',equ)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
